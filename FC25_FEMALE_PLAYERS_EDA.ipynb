# Analysing Player Attributes

Following analysis involves:

- Distribution of each player attribute
- Correlation between each attribute
- How these attributes have an affect on a player's overall rating

```python
#distribution of columns

print("\nDistribution of Numeric Features")
dff.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
print("\nDistribution of Player Age")
dff.hist(column=['Age'], figsize=(4,4))
plt.tight_layout()
plt.show()
```

```python
print("\nDistribution of Main Stats")
dff.hist(column=['OVR', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY'], figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
print("\nDistribution of Goalkeeper Main Stats")
dff.hist(column=['GK Diving', 'GK Handling', 'GK Kicking', 'GK Positioning', 'GK Reflexes'], figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix between different player attributes
corr_matrix = dff[['OVR', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY']].corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='RdYlGn', linewidths=0.5, mask = mask)
plt.title('Correlation Matrix of Player Attributes')
plt.show()
```

```python
# Correlation matrix for Goalkeepers
corr_matrix = dff[['GK Diving', 'GK Handling', 'GK Kicking', 'GK Positioning', 'GK Reflexes']].corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='RdYlGn', linewidths=0.5, mask = mask)
plt.title('Correlation Matrix of Player Attributes')
plt.show()
```

```python
# Player attributes impact on overall rating

attributes = ['PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY']

figure, ax = plt.subplots(2, 3, figsize=(10, 6))
figure.suptitle('Player Attributes Impact on Overall Rating')
for i, attribute in enumerate(attributes):
    sns.regplot(x='OVR', y=attribute, data=dff, ax=ax[i//3, i%3], marker = '.', color='blue', line_kws=dict(alpha=0.8, color='red'))
    ax[i//3, i%3].set_title(f'{attribute} vs OVR')
    ax[i//3, i%3].set_xlabel('OVR')
    ax[i//3, i%3].set_ylabel(attribute)
plt.tight_layout()
plt.show()
```

```python
# Age impact on overall rating

plt.figure(figsize=(10, 6))
sns.regplot(x='OVR', y='Age', data=dff, color='blue', marker='.', line_kws=dict(alpha=0.8, color='red'), x_jitter=0.5,
            y_jitter=0.5)
plt.title('Age vs Overall Rating')
plt.grid(True, linestyle='-', alpha=0.6)
plt.xlabel('OVR')
plt.ylabel('Age')
plt.show()
```

# Analysis on the Women Leagues in EA FC25

Following analysis involves:

- Average rating of each league in the game
- Amount of players in each league
- Unique leagues and what overall the highest rated player is

```python
# Top 10 Highest Rated Leagues

top_leagues = dff.groupby('League')['OVR'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_leagues.values, y=top_leagues.index, palette='RdYlBu')
plt.title('Top 10 Highest Rated Leagues')
for x, y in enumerate(top_leagues.values):
    plt.text(y + 0.5, x, f'{y:.1f}', ha='left', color='black')
plt.xlabel('Average OVR')
plt.ylabel('League')
plt.show()
```

```python
# Group by league and calculate the average overall rating
league_avg_ovr = dff.groupby(['League'])['OVR'].mean().sort_values(ascending=False).reset_index()

# Plotting the ranked leagues based on their average OVR
plt.figure(figsize=(12, 10))  # Increase the figure height if needed
bar_plot = sns.barplot(x='OVR', y='League', data=league_avg_ovr, palette='RdYlBu')

# Increase bar width
for bar in bar_plot.patches:
    bar.set_height(0.6)  # Adjust the width as needed

# Add title and labels
plt.title('Leagues Ranked by Average Overall Rating (OVR)', fontsize=16)
plt.xlabel('Average Overall Rating (OVR)')
plt.xlim([50,80])
plt.ylabel('League')

# Add value labels on each bar
for index, (league, ovr) in enumerate(zip(league_avg_ovr['League'], league_avg_ovr['OVR'])):
    plt.text(ovr + 0.5, index, f'{ovr:.2f}', va='center', ha='left', fontsize=10)

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```

```python
# Count of players in top leagues

top_leagues = ["Calcio A Femminile", "Barclays WSL", "Arkema PL", "NWSL", "GPFBL", "Liga F"]

filtered_df = dff.loc[dff['League'].isin(top_leagues)]

league_player_counts = filtered_df['League'].value_counts()

print("\nCountplot of Players in Top league's (sorted by count)")
plt.figure(figsize=(10, 6))
sns.countplot(y='League', data=filtered_df, palette='Set1', order=league_player_counts.index)
plt.title('Count of FC25 Female Players in each League')
plt.tight_layout()
plt.show()
```

```python
# Count of players in each league

league_player_counts = dff['League'].value_counts().index.tolist()

print("\nCountplot of Players in each league (sorted by count)")
plt.figure(figsize=(12, 10))
sns.countplot(y='League', data=dff, palette='Set1', order=league_player_counts)
plt.title('Count of FC25 Female Players in each League')
plt.tight_layout()
plt.show()
```

```python
# All leagues and their highest rated players

highest_per_team = dff.loc[dff.groupby('League')['OVR'].idxmax()][['League', 'Name', 'OVR']]
highest_per_team.reset_index(drop=True, inplace=True)
highest_per_team.sort_values('OVR', ascending=False).reset_index(drop=True)
```

# Analysis on the Teams in EA FC25

Following analysis involves:

- Highest rated teams and their highest rated player
- Average rating of Top leagues

```python
# Top 10 highest rated teams

top_teams = dff.groupby('Team')['OVR'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_teams.values, y=top_teams.index, palette='RdYlBu')
plt.title('Top 10 Highest Rated Teams')
for x, y in enumerate(top_teams.values):
    plt.text(y + 0.5, x, f'{y:.1f}', ha='left', color='black', va='center')
plt.xlabel('Average OVR')
plt.ylabel('Teams')
plt.show()
```

# Barclays WSL

```python
# Top 10 highest rated Barclays WSL teams

barclays_wsl_teams = dff[dff['League'] == 'Barclays WSL']
top_teams_wsl = barclays_wsl_teams.groupby('Team')['OVR'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(14,8))
sns.scatterplot(x=top_teams_wsl.values, y=top_teams_wsl.index)
plt.title('Top 10 Highest Rated Barclays WSL Teams')
for x, y in enumerate(top_teams_wsl.values):
    plt.annotate(f'{y:.1f}', (y + 0.2, x), ha='center', va='center', fontsize=12)
plt.xlabel('Average OVR')
plt.ylabel('Teams')
plt.show()
```

```python
# Rating gap between highest and lowest rated team in the Barclays WSL
barclays_wsl_data = dff[dff['League'] == 'Barclays WSL']

# Group by team and calculate average OVR
team_averages = barclays_wsl_data.groupby('Team')['OVR'].mean()

# Find maximum and minimum average OVRs for Premier League teams
max_avg_ovr_wsl = team_averages.max()
min_avg_ovr_wsl = team_averages.min()

# Print the results
print("Maximum average OVR for Barclays WSL:", max_avg_ovr_wsl)
print("Minimum average OVR for Barclays WSL:", min_avg_ovr_wsl)
```
