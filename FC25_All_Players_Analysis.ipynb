# Comparative Data Analysis on the Male and Female Players on EA FC25

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

dfa=pd.read_csv('all_players.csv')
dfa.head()
```

# Analysis comparing the male and female leagues in EA FC25

```python
# Top 10 Highest Rated Leagues

top_leagues = dfa.groupby('League')['OVR'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_leagues.values, y=top_leagues.index, palette='RdYlBu')
plt.title('Top 10 Highest Rated Leagues')
for x, y in enumerate(top_leagues.values):
    plt.text(y + 0.5, x, f'{y:.1f}', ha='left', color='black')
plt.xlabel('Average OVR')
plt.ylabel('League')
plt.show()
```

```python
# Count of players in top leagues

top_leagues = ["Calcio A Femminile", "Premier League", "Barclays WSL", "Arkema PL", "LALIGA EA SPORTS"]

filtered_df = dfa.loc[dfa['League'].isin(top_leagues)]

league_player_counts = filtered_df['League'].value_counts()

print("\nCountplot of Players in Top league's (sorted by count)")
plt.figure(figsize=(10, 6))
sns.countplot(y='League', data=filtered_df, palette='Set1', order=league_player_counts.index)
plt.title('Count of FC25 Players in each League')
plt.tight_layout()
plt.show()
```

```python
# top 10 highest rated leagues and their highest rated players

highest_per_team = dfa.loc[dfa.groupby('League')['OVR'].idxmax()][['League', 'Name', 'OVR']]
highest_per_team.reset_index(drop=True, inplace=True)
highest_per_team.sort_values('OVR', ascending=False).head(10).reset_index(drop=True)
```

# Analysis on all Teams in EA FC25

```python
# top 10 highest rated teams and their highest rated players

highest_per_team = dfa.loc[dfa.groupby('Team')['OVR'].idxmax()][['Team', 'Name', 'OVR']]
highest_per_team.reset_index(drop=True, inplace=True)
highest_per_team.sort_values('OVR', ascending=False).head(10).reset_index(drop=True)
```

```python
# Top 10 highest rated teams
#Lombardia = Inter Milan

top_teams = dfa.groupby('Team')['OVR'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_teams.values, y=top_teams.index, palette='RdYlBu')
plt.title('Top 10 Highest Rated Teams')
for x, y in enumerate(top_teams.values):
    plt.text(y + 0.5, x, f'{y:.1f}', ha='left', color='black', va='center')
plt.xlabel('Average OVR')
plt.ylabel('Teams')
plt.show()
```

```python
top_leagues_min_max = ['Premier League', "Calcio A Femminilie", 'Barclays WSL', 'LALIGA EA SPORTS', 'Arkema PL', 'Serie A Enilive']
filtered_df = dfa[dfa['League'].isin(top_leagues_min_max)]

# Group by league and calculate max and min average OVR
league_stats = filtered_df.groupby('League')['OVR'].agg(['max', 'min'])

# Create the bar plot 
league_stats.plot(kind='bar')
plt.title('Max and Min Average OVR by League')
plt.ylabel('Average OVR')
plt.xticks(rotation=45)
plt.show()
```

# Analysis on players in EA FC25

```python
# Top 10 Players with Highest Rating
ovr_rank = dfa.sort_values('OVR', ascending=False)
overall_rank = ovr_rank[['Name', 'Age', 'Nation', 'Team', 'OVR']].head(10)
overall_rank
```

```python
# Top 10 players visualization with barplot

plt.figure(figsize=(10, 6))
plt_bar = sns.barplot(x='OVR', y='Name', data=overall_rank, palette='flare')
for bar in plt_bar.patches:
    bar.set_height(0.8)
plt.title('Top 10 Players by Overall Rating')
for index, value in enumerate(overall_rank['OVR']):
    plt.text(value + 0.5, index, str(value), ha='left', va='center')
plt.xlabel('OVR')
plt.ylabel('Name')
plt.show()
```
