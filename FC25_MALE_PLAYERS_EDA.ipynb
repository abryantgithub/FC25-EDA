# Analysing Player Attributes

Following analysis involves:

- Distrubution of each player attribute
- Correlation between each attribute
- How these attributes have an affect on a player's overall rating

```python
#distribution of columns

print("\nDistribution of Numeric Features")
dfm.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
print("\nDistribution of Player Age")
dfm.hist(column=['Age'], figsize=(4,4))
plt.tight_layout()
plt.show()
```

```python
print("\nDistribution of Main Stats")
dfm.hist(column=['OVR', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY'], figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
print("\nDistribution of Goalkeeper Main Stats")
dfm.hist(column=['GK Diving', 'GK Handling', 'GK Kicking', 'GK Positioning', 'GK Reflexes'], figsize=(10, 8))
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix between different player attributes
corr_matrix = dfm[['OVR', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY']].corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='RdYlGn', linewidths=0.5, mask = mask)
plt.title('Correlation Matrix of Player Attributes')
plt.show()
```

```python
# Correlation matrix for Goalkeepers
corr_matrix = dfm[['GK Diving', 'GK Handling', 'GK Kicking', 'GK Positioning', 'GK Reflexes']].corr()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='RdYlGn', linewidths=0.5, mask = mask)
plt.title('Correlation Matrix of Player Attributes')
plt.show()
```

```python
# Player attributes impact on overall rating

attributes = ['PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY']

figure, ax = plt.subplots(2, 3, figsize=(10, 6))
figure.suptitle('Player Attributes Impact on Overall Rating')
for i, attribute in enumerate(attributes):
    sns.regplot(x='OVR', y=attribute, data=dfm, ax=ax[i//3, i%3], marker = '.', color='blue', line_kws=dict(alpha=0.8, color='red'))
    ax[i//3, i%3].set_title(f'{attribute} vs OVR')
    ax[i//3, i%3].set_xlabel('OVR')
    ax[i//3, i%3].set_ylabel(attribute)
plt.tight_layout()
plt.show()
```

```python
# Age impact on overall rating

plt.figure(figsize=(10, 6))
sns.regplot(x='OVR', y='Age', data=dfm, color='blue', marker='.', line_kws=dict(alpha=0.8, color='red'), x_jitter=0.5,
            y_jitter=0.5)
plt.title('Age vs Overall Rating')
plt.grid(True, linestyle='-', alpha=0.6)
plt.xlabel('OVR')
plt.ylabel('Age')
plt.show()
```

# Analysis on the Leagues in EA FC25

Following analysis involves:

- Average rating of each league in the game
- Amount of players in each league
- 10 unique leagues and what overall the highest rated player is

```python
# Top 10 Highest Rated Leagues

top_leagues = dfm.groupby('League')['OVR'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_leagues.values, y=top_leagues.index, palette='RdYlBu')
plt.title('Top 10 Highest Rated Leagues')
for x, y in enumerate(top_leagues.values):
    plt.text(y + 0.5, x, f'{y:.1f}', ha='left', color='black')
plt.xlabel('Average OVR')
plt.ylabel('League')
plt.show()
```

```python
# Group by league and calculate the average overall rating
league_avg_ovr = dfm.groupby(['League'])['OVR'].mean().sort_values(ascending=False).reset_index()

# Plotting the ranked leagues based on their average OVR
plt.figure(figsize=(12, 10))  # Increase the figure height if needed
bar_plot = sns.barplot(x='OVR', y='League', data=league_avg_ovr, palette='RdYlBu')

# Increase bar width
for bar in bar_plot.patches:
    bar.set_height(0.6)  # Adjust the width as needed

# Add title and labels
plt.title('Leagues Ranked by Average Overall Rating (OVR)', fontsize=16)
plt.xlabel('Average Overall Rating (OVR)')
plt.xlim([50,80])
plt.ylabel('League')

# Add value labels on each bar
for index, (league, ovr) in enumerate(zip(league_avg_ovr['League'], league_avg_ovr['OVR'])):
    plt.text(ovr + 0.5, index, f'{ovr:.2f}', va='center', ha='left', fontsize=10)

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```

```python
# Count of players in top leagues

top_leagues = ["Premier League", "LALIGA EA SPORTS", "Serie A Enilive", "Bundesliga", "Ligue 1 McDonald's"]

filtered_df = dfm.loc[dfm['League'].isin(top_leagues)]

league_player_counts = filtered_df['League'].value_counts()

print("\nCountplot of Players in Top league's (sorted by count)")
plt.figure(figsize=(10, 6))
sns.countplot(y='League', data=filtered_df, palette='Set1', order=league_player_counts.index)
plt.title('Count of FC25 Male Players in each League')
plt.tight_layout()
plt.show()
```

```python
# Count of players in each league

league_player_counts = dfm['League'].value_counts().index.tolist()

print("\nCountplot of Players in each league (sorted by count)")
plt.figure(figsize=(12, 10))
sns.countplot(y='League', data=dfm, palette='Set1', order=league_player_counts)
plt.title('Count of FC25 Male Players in each League')
plt.tight_layout()
plt.show()
```

```python
# top 10 highest rated leagues and their highest rated players

highest_per_team = dfm.loc[dfm.groupby('League')['OVR'].idxmax()][['League', 'Name', 'OVR']]
highest_per_team.reset_index(drop=True, inplace=True)
highest_per_team.sort_values('OVR', ascending=False).head(10).reset_index(drop=True)
```

# Analysis on Player Positions

Following analysis involves:

- Rating by position
- Attributes by position
- Amount of players in each position

```python
positions = dfm.groupby('Position')['OVR'].median().sort_values().index
plt.figure(figsize=(10, 6))

sns.boxplot(x='Position',y='OVR', data = dfm, order = positions)
plt.title('Overall rating by position')
plt.ylabel('Overall rating')
plt.show()
```

```python
attributes = ['PAC', 'SHO', 'DRI', 'PHY', 'DEF', 'PAS']
# Set up the figure and axes for subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(18, 15), sharey=True,)
fig.suptitle('Player attributes by position', fontsize=16)

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot the distribution of each attribute in a separate subplot
for i, attr in enumerate(attributes):
    order = dfm.groupby('Position')[attr].median().sort_values().index
    
    # Plot the boxplot with ordered categories
    sns.boxplot(x='Position', y=attr, data=dfm, ax=axes[i], order=order, palette = 'PiYG')
    axes[i].set_title(f'{attr} rating by position')
    axes[i].set_ylabel(f'{attr}')
    axes[i].set_xlabel('Position')
# Adjust the layout to prevent overlap
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
# Count of players in each position

position_counts = dfm['Position'].value_counts().index.tolist()

print("\nCountplot of Player positions (sorted by count)")
plt.figure(figsize=(10, 6))
sns.countplot(y='Position', data=dfm, palette='Set1', order=position_counts)
plt.title('Count of FC25 Male Player Positions')
plt.show()
```

```python
# Now grouping positions into GK, DEF, MID, ATT

m_position_groups = {
    "GK": ["GK"],  # Goalkeepers
    "DEF": ["CB", "LB", "RB"],  # Defenders 
    "MID": ["CDM", "CM", "CAM", "LM", "RM"],  # Midfielders 
    "FW": ["ST", "LW", "RW"],  # Attackers 
}

# Create a new column with group names based on the original position
dfm['Position_Group'] = dfm['Position'].apply(lambda p: next((g for g, positions in m_position_groups.items() if p in positions), None))

# Handle positions not in the mapping (optional)
dfm['Position_Group'].fillna('Other', inplace=True)  # Fill missing values with 'Other'

# Calculate group counts and sort
group_counts = dfm['Position_Group'].value_counts().sort_values(ascending=False)

print("\nCountplot of Player Positions (Grouped)")
plt.figure(figsize=(10, 6))
sns.countplot(y='Position_Group', data=dfm, palette='Set1', order=group_counts.index)
plt.title('Count of FC25 Male Player Positions (Grouped)')
plt.show()
```
